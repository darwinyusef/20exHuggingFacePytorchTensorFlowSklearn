```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread("/content/mi_archivo.JPG", cv2.IMREAD_GRAYSCALE)

if imagen is None:
    raise ValueError("Error: No se pudo cargar la imagen. Verifica la ruta.")

# Definir kernels personalizados
mis_kernels = {
    "kernelUnito": np.array([[1, -1, 1], [-1, 5, -1], [1, -1, 1]]),
    "kernelDosito": np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]]),
    "kernelTresito": np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]),
    "kernelCuatrito": np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]]) # Sobel X modificado
}

# Aplicar los kernels y mostrar resultados
plt.figure(figsize=(12, 6))
plt.subplot(1, len(mis_kernels) + 1, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

for i, (nombre, kernel) in enumerate(mis_kernels.items(), start=2):
    imagen_filtrada = cv2.filter2D(imagen, -1, kernel)
    plt.subplot(1, len(mis_kernels) + 1, i)
    plt.imshow(imagen_filtrada, cmap="gray")
    plt.title(nombre)
    plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Cargar la imagen en escala de grises
imagen = cv2.imread("/content/mi_archivo.JPG", cv2.IMREAD_GRAYSCALE)

if imagen is None:
    raise ValueError("Error: No se pudo cargar la imagen. Verifica la ruta.")

# Kernel base
kernel_base = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])

# Número de incrementos
num_incrementos = +9  # Puedes cambiar este valor

# Generar una lista de kernels con valores incrementados
kernels_incrementados = {
    f"Kernel +{i}": kernel_base + i for i in range(num_incrementos)
}

# Mostrar imagen original
plt.figure(figsize=(12, 6))
plt.subplot(1, num_incrementos + 1, 1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

# Aplicar cada kernel y mostrar el resultado
for i, (nombre, kernel) in enumerate(kernels_incrementados.items(), start=2):
    imagen_filtrada = cv2.filter2D(imagen, -1, kernel)
    plt.subplot(1, num_incrementos + 1, i)
    plt.imshow(imagen_filtrada, cmap="gray")
    plt.title(nombre)
    plt.axis("off")

plt.show()
```
